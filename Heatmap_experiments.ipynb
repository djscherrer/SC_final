```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

fields = ['Computer Science', 'Psychology', 'Biology', 'History', 'Linguistics']
universities = {
    "Switzerland": ["ETH Zurich", "University of Geneva", "University of Zurich"],
    "UK": ["University of Oxford", "University of Cambridge", "Imperial College London"],
    "China": ["Tsinghua University", "Peking University", "Fudan University"],
    "India": ["Indian Institute of Science", "Indian Institute of Technology Delhi", "Indian Institute of Technology Bombay"],
    "Iran": ["University of Tehran", "Sharif University of Technology", "Amirkabir University of Technology"],
    "USA": ["Massachusetts Institute of Technology", "Stanford University", "Harvard University"]
}

universities = [university for univs in universities.values() for university in univs]


# Create a DataFrame with random values between 1 and 10
data = np.random.randint(3, 11, size=(len(fields), len(universities)))
df = pd.DataFrame(data, columns=universities, index=fields)
df = df[df.sum().sort_values(ascending=False).index]

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='viridis', fmt='d', vmin=1, vmax=10)
plt.title('Overall Quality Bias of Universities in Different Fields, Perceived by ChatGPT')
plt.ylabel('Research Fields')
plt.show()

```
